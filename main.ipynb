```python
from nltk.corpus import brown

vocab = []
for word in set(brown.words()):
  if(word.isalpha()):
    vocab.append(word)
  
print(len(vocab))
```

```python
import re
correct_words = []
statements = []

with open("raw.txt", "r") as f:
  all_lines = f.readlines()
  for line in all_lines:
    word = re.findall("\A[a-zA-Z]*", line)[0]
    statement = line.replace(word, "").replace("*", "").strip()

    correct_words.append(word)
    statements.append(statement)

statements = statements[:10]
```

```python
from modeloutput import model_output
from successatk import success_at_k

models = [1,2,3]
k_values = [1,5,10]
result = {}
run = {}
for n in models:
  temp_res_1 = {}
  temp_run_1 = {}
  for i in range(len(statements)):
    temp_res_2 = {}
    temp_run_2 = {}
    for k in k_values:
      temp_res_2[k] = 0
      temp_run_2[k] = 1
    temp_res_1[statements[i]] = temp_res_2
    temp_run_1[statements[i]] = temp_run_2
  result[n] = temp_res_1
  run[n] = temp_run_1

for n in models:
  index = -1 * n if n > 1 else -2
  model_name = str("LMn-" + str(n))
  for i in range(len(statements)):
    print("N: ", n)
    res_list = model_output(model_name, vocab, statements[i].split()[index+1:])
    print("Result list : ",res_list)
    for k in k_values:
      result[n][statements[i]][k] = success_at_k(correct_words[i], res_list, k)
      # run[n][statements[i]][k] = 1
      print("N : ", n, "\t\tStatement: ", statements[i].split()[index+1:], "\t\tWord: ", correct_words[i], "\t\tK : ", k)
```
